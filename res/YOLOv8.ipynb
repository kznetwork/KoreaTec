## 코드 개선

```python
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Load a pretrained YOLO model
model = YOLO("yolov8n.pt")  # COCO dataset으로 pretrained된 모델을 불러옴

# Define image file path
image_path = "C:\\DEV\\KoreaTech\\YOLO\\im1.jpg"

# Use the model to perform object detection
results = model.predict(image_path)

# Print results for debugging
print(results[0])
print()
print(results[0].boxes)  # 좌상단 좌표, 우하단 좌표, confidence score, class id

# Plot the result
res_plotted = results[0].plot()  # 결과 이미지를 그립니다 (bounding box 등)

# Convert BGR (OpenCV format) to RGB (matplotlib format)
res_plotted_rgb = cv2.cvtColor(res_plotted, cv2.COLOR_BGR2RGB)

# Display the image using matplotlib
plt.figure(figsize=(12, 12))
plt.imshow(res_plotted_rgb)
plt.axis('off')  # 축을 숨깁니다
plt.show()
```

## YOLOv8의 예측 결과를 그림 위에 "탐지한 객체 수"를 노란색 글씨로 표시

```python
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO

# 1. Load pretrained YOLO model
model = YOLO("yolov8n.pt")

# 2. Load image
image_path = "C:\\DEV\\KoreaTech\\YOLO\\im1.jpg"

# 3. Object detection
results = model.predict(image_path)
result = results[0]
boxes = result.boxes

# 4. Count detected objects
num_objects = len(boxes)

# 5. Draw results on image
res_plotted = result.plot()  # OpenCV 이미지 (BGR)
text = f"Detected objects: {num_objects}"

# 6. Put yellow text on image (좌상단에 표시)
cv2.putText(
    res_plotted,                     # 이미지
    text,                            # 출력할 텍스트
    (30, 50),                        # 텍스트 위치 (x, y)
    cv2.FONT_HERSHEY_SIMPLEX,       # 폰트
    1.3,                             # 폰트 크기 (13pt 정도에 해당)
    (0, 255, 255),                   # 노란색 (BGR)
    2,                               # 두께
    cv2.LINE_AA                      # 안티앨리어싱
)

# 7. Convert BGR to RGB for matplotlib
res_plotted_rgb = cv2.cvtColor(res_plotted, cv2.COLOR_BGR2RGB)

# 8. Show the image
plt.figure(figsize=(12, 12))
plt.imshow(res_plotted_rgb)
plt.axis('off')
plt.show()
```
